```python
import pandas as pd

df=pd.read_csv('SMSSpamCollection.csv',header=None,error_bad_lines=False)

df.head()
```

```python
df.rename(columns = {0:'Label'}, inplace = True) 
df.rename(columns = {1:'Text'}, inplace = True) #rename column names

df=df.drop(df.index[0])
df
```

```python
df=df.dropna()
```

```python

X=df.drop('Label',axis=1)
```

```python

## Get the Dependent features
y=df['Label']
```

```python
X.shape
```

```python
y.shape
```

```python

import tensorflow as tf
```

```python

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
#we need to add 0 in front or back to make sentence equal in len
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python

### Vocabulary size any number
voc_size=5000
```

```python
messages=X.copy()
```

```python
messages['Text']
```

```python
messages.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['Text'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    #if word in review not in stop wordsthen only stem itand put back to review
    review = ' '.join(review)
    #then join the words in review
    corpus.append(review)
    #and form sentences
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus]
```

```python
onehot_repr
#each word in a sentence is represented as index out of 5000
```

```python
#before giving to embedding layer make all sentences of equal len
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)#post=add at end
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
## Creating model
embedding_vector_features=40 #any
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))#sentlen=20
#output of embedding layer is given to lstm layer
model.add(LSTM(100))#one lstm with 100 neuron
model.add(Dense(1,activation='sigmoid'))#classification prob
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python

X_final.shape,y_final.shape
```

```python
X_final
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
### Finally Training
model.fit(X_train,y_train,epochs=10,batch_size=64)
```

```python
#accuracy
y_pred=model.predict_classes(X_test)
y_pred
```

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```
